```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
import gradio as gr

# .env íŒŒì¼ ë¡œë“œ
load_dotenv()

# OpenAI LLM ì¤€ë¹„
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.5)
```

## í”„ë¡¬í”„íŠ¸ - ì‹œìŠ¤í…œ ì—­í•  ë©”ì‹œì§€ ë° ì‚¬ìš©ì ë©”ì‹œì§€ êµ¬ì„±

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "ë‹¹ì‹ ì€ 'ëª…íƒì • ì½”ë‚œ' ì „ë¬¸ê°€ì…ë‹ˆë‹¤. ëª¨ë“  ì§ˆë¬¸ì— ì¹œì ˆí•˜ê³  ì •í™•í•˜ê²Œ ë‹µí•´ì£¼ì„¸ìš”."),
    ("human", "{question}")
])
```

## ì²´ì¸ êµ¬ì„± ë° UI ì…ì¶œë ¥ í•¨ìˆ˜ ì •ì˜

### LCEL ì²´ì¸ êµ¬ì„± 

ì²´ì¸ (í”„ë¡¬í”„íŠ¸ â†’ LLM â†’ ì¶œë ¥ íŒŒì„œ)ë¥¼ íŒŒì´í”„(|)ë¡œ ì—°ê²°
- `ChatPromptTemplate`: ë¬¸ìì—´ ë©”ì‹œì§€ ìƒì„±
- `LLM`: ì‘ë‹µ ìƒì„±
- `StrOutputParser`: LLM ì‘ë‹µì—ì„œ í…ìŠ¤íŠ¸ ì¶”ì¶œ

### Gradio ì…ì¶œë ¥ í•¨ìˆ˜ ì •ì˜
- `conan_expert_reply` : ì‚¬ìš©ì ì…ë ¥ê°’ì„ ë°›ì•„ LCEL ì²´ì¸ì— ì „ë‹¬í•˜ê³  ê²°ê³¼ ë¦¬í„´


```python
# LCEL ì²´ì¸ êµ¬ì„±
chain = prompt | llm | StrOutputParser()

# Gradio ì…ì¶œë ¥ í•¨ìˆ˜ ì •ì˜
def conan_expert_reply(user_input):
    return chain.invoke({"question": user_input})
```

## Gradio UI êµ¬ì„±

```python
with gr.Blocks() as demo:
    gr.Markdown("### ğŸ•µï¸â€â™‚ï¸ ëª…íƒì • ì½”ë‚œ ì „ë¬¸ê°€ (ì§ˆì˜ & ì‘ë‹µ)")

    # ì…/ì¶œë ¥ ì˜ì—­ êµ¬ì„±
    question = gr.Textbox(label="ì§ˆë¬¸ì„ ì…ë ¥í•˜ì„¸ìš”", placeholder="ì½”ë‚œì´ ì‘ì•„ì§„ ì´ìœ ëŠ” ë­ì£ ?")
    answer = gr.Textbox(label="ì½”ë‚œ ì „ë¬¸ê°€ì˜ ë‹µë³€", lines=5)

    # ì…ë ¥ ì´ë²¤íŠ¸ ì²˜ë¦¬
    question.submit(fn=conan_expert_reply, inputs=question, outputs=answer)

# Gradio ì‹¤í–‰
demo.launch()
```
